```python
# FROM: http://scikit-learn.org/stable/auto_examples/svm/plot_svm_kernels.html#example-svm-plot-svm-kernels-py

import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
```

```python
# Our dataset and targets
X = np.c_[
          # 0
          (.4, -.7),
          (-1.5, -1),
          (-1.4, -.9),
          (-1.3, -1.2),
          (-1.1, -.2),
          (-1.2, -.4),
          (-.5, 1.2),
          (-1.5, 2.1),
          (1, 1),

          # 1
          (1.3, .8),
          (1.2, .5),
          (.2, -2),
          (.5, -2.4),
          (.2, -2.3),
          (0, -2.7),
          (1.3, 2.1)].T

Y = [0] * 8 + [1] * 8
```

```python
fignum = 1

x_min = -3
x_max = 3
y_min = -3
y_max = 3

for kernel in ('linear', 'poly', 'rbf'):
    clf = svm.SVC(kernel=kernel)
    clf.fit(X, Y)
    plt.figure(fignum, figsize=(4, 3))
    plt.clf()

#     plt.scatter(X[:, 0], X[:, 1], c=Y, zorder=10, cmap=plt.cm.Paired)
#     plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=80, facecolors='none', zorder=10)

    plt.axis('tight')
    
    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
    
    Z = clf.decision_function(np.c_[XX.ravel(), YY.ravel()])

    Z = Z.reshape(XX.shape)
    
    plt.figure(fignum, figsize=(4, 3))
    plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
    plt.contour(XX, YY, Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'], levels=[-1, 0, 1])
    
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    
    fignum = fignum + 1
    
plt.show()
```
