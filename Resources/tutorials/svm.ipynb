```python

def compare_svc_kernels(X, y, C=1.0):
    """
    Create a instances of SVM classifiers with different Kernels and plot the decision 
    boundary given our fitted our data. 
    
    We do not scale our data since we want to plot the support vectors.
    """

    svc = svm.SVC(kernel='linear', C=C).fit(X, y)
    rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(X, y)
    poly_svc = svm.SVC(kernel='poly', degree=3, C=C).fit(X, y)
    lin_svc = svm.LinearSVC(C=C, multi_class='ovr').fit(X, y)
    

    # title for the plots
    titles = ['SVC with linear kernel',
              'LinearSVC (linear kernel)',
              'SVC with RBF kernel',
              'SVC with polynomial (degree 3) kernel']

    plt.subplots(2, 2, figsize=(20, 20))

    for i, clf in enumerate((svc, lin_svc, rbf_svc, poly_svc)):
        plt.subplot(2, 2, i + 1)
        
        plot_decision_boundary(clf, X, y, title=titles[i])


    plt.show()

    
    
# Read each line of code in the following function and add your own comments
def plot_decision_boundary(classifier, X, y, density=False, title=None):
    x0 = X[:, 0]
    x1 = X[:, 1]
    
    x_min = x0.min() - 0.1 * np.abs(x0.min())
    x_max = x0.max() + 0.1 * np.abs(x0.max())
    
    y_min = x1.min() - 0.1 * np.abs(x1.min())
    y_max = x1.max() + 0.1 * np.abs(x1.max())

    x_range = np.linspace(x_min, x_max, 100)
    y_range = np.linspace(y_min, y_max, 100)

    xx, yy = np.meshgrid(x_range, y_range)
    zz = classifier.decision_function(np.c_[xx.ravel(), yy.ravel()])
    zz = zz.reshape(xx.shape)

    plt.pcolormesh(xx, yy, zz > 0, cmap=plt.cm.Paired, alpha = .2, edgecolors='None', antialiased=True);
    plt.contour(xx, yy, zz, colors=['r', 'g', 'r'], linestyles=['--', '-', '--'], levels=[-1, 0, 1], linewidths=[3, 6, 3])
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, s=80, alpha=0.5);
    plt.gca().set_title(title)

    
    
def plot_train_test(classifier, X_train, X_test, y_train, y_test):
    plt.subplots(1, 2, figsize=(18, 5))

    plt.subplot(121)
    axes = plt.gca()
    axes.set_title("Training Data")
    plot_decision_boundary(classifier, X_train, y_train)

    plt.subplot(122)
    axes = plt.gca()
    axes.set_title("Testing Data")
    plot_decision_boundary(classifier, X_test, y_test)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn import svm
from sklearn import metrics
from sklearn.cross_validation import train_test_split

%matplotlib inline

iris_data = datasets.load_iris()

iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
```

## Weighted classes

```python
# we create 40 separable points
rng = np.random.RandomState(0)
n_samples_1 = 1000
n_samples_2 = 100
X = np.r_[1.5 * rng.randn(n_samples_1, 2),
          0.5 * rng.randn(n_samples_2, 2) + [2, 2]]
y = [0] * (n_samples_1) + [1] * (n_samples_2)

# fit the model and get the separating hyperplane
clf = svm.SVC(kernel='linear', C=1.0)
clf.fit(X, y)

w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-5, 5)
yy = a * xx - clf.intercept_[0] / w[1]

# get the separating hyperplane using weighted classes
wclf = svm.SVC(kernel='linear', class_weight={1: 10})

wclf.fit(X, y)

ww = wclf.coef_[0]
wa = -ww[0] / ww[1]
wyy = wa * xx - wclf.intercept_[0] / ww[1]

# plot separating hyperplanes and samples
plt.figure(figsize=(9, 9))
h0 = plt.plot(xx, yy, 'k-', label='no weights')
h1 = plt.plot(xx, wyy, 'k--', label='with weights')

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.legend()

plt.axis('tight')
plt.show()
```
