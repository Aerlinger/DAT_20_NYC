```python
import pandas as pd
import datetime
from pandas.tseries.offsets import *

import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split

exp = pd.read_csv('C:\Dan Temp\DataScienceProjData.csv')

exp['TransDate'] = pd.to_datetime(exp['TransDate'])
exp['Month'] = exp['TransDate'].map(lambda x: 12*(x.year-2006) + x.month)

exp
```

```python
sum_exp = exp[exp.Medium !='Transfer']
sum_exp = exp.groupby(['Month', 'Medium','Relationship','DisposableIncome','City'])['TransAmount'].sum().reset_index()
#pivoted = sum_exp.pivot(['Month','Relationship'], 'Medium')
pivoted = pd.pivot_table(sum_exp, values='TransAmount', index=['Month', 'Relationship','DisposableIncome','City'], columns=['Medium']).reset_index().fillna(0)
#pivoted = pivoted.drop('Transfer', axis=1, inplace=True)
pivoted
#print sum_exp
```

```python
# create X and y
feature_cols = ['Month','Relationship', 'DisposableIncome']
X = pivoted[feature_cols]
y = pivoted.Food

# follow the usual sklearn pattern: import, instantiate, fit
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X, y)

# print intercept and coefficients
print lm.intercept_
print lm.coef_
```

```python
X = exp[['Month', 'DisposableIncome','TransAmount']]
y = exp.Relationship
# random_state is just a seed value for the initial state
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Convert them back into dataframes, for convenience
train = pd.DataFrame(data=X_train, columns=['Month', 'DisposableIncome','TransAmount'])
train['Relationship'] = y_train

test = pd.DataFrame(data=X_test, columns=['Month', 'DisposableIncome','TransAmount'])
test['Relationship'] = y_test



balance = LogisticRegression()
balance.fit(train[['TransAmount']], y_train)
B1 = balance.coef_[0][0]
B0 = balance.intercept_[0]
np.exp(B1)

print "Beta1", B1
print "Beta0", B0
```

```python
test['pred_class'] = balance.predict(test[['TransAmount']])

# 5 - Compute the overall accuracy, the sensitivity and specificity
# Accuracy
accuracy = sum(test.pred_class == test.Relationship) / float(len(test.Relationship))

# Specificity
# For those who didn't default, how many did it predict correctly?
test_nd = test[test.Relationship == 0]
specificity = sum(test_nd.pred_class == test_nd.Relationship) / float(len(test_nd.Relationship))

# Sensitivity
# For those who did default, how many did it predict correctly? 
test_d = test[test.Relationship == 1]
sensitivity = sum(test_d.pred_class == test_d.Relationship) / float(len(test_d.Relationship))

# This raises the question, how does our overall 
# classification accuracy compare to the not-default rate?
#null = 1 - sum(d.Relationship) / float(len(d.Relationship))

# This illustrates an important point, class imbalance can result in accuracy
# measures that are misleading. After all, if you would have just guessed not
# going to default, you would be correct 96.67 % of the time.

f1_score = 2 * (sensitivity * specificity) / (sensitivity + specificity)

print accuracy
print "Sensitivity (Aka Recall proportion of actual positives which are correctly - identified): ", sensitivity
print "Specificity (proportion of negatives which are correctly identified): ", specificity

print "F1 Score: ", f1_score
```
