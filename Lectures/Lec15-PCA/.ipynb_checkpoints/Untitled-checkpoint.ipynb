```python
from sklearn import datasets

iris = datasets.load_iris()

df = pd.DataFrame(iris.data)
df.columns=['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid']
df.dropna(how="all", inplace=True) # drops the empty line at file-end


pd.scatter_matrix(df, figsize=(10, 10));
```

```python
plt.scatter(df.petal_wid, df.petal_len)
from mpl_toolkits.mplot3d import Axes3D
 
```

```python
from sklearn.decomposition import PCA

X = df

pca = PCA()

pca.fit(X)
```

```python
components = pca.components_
pca_score = pca.explained_variance_ratio_

components
```

```python
pca_score

components.T * pca_score / pca_score.min()
```

```python
x1_pca_axis, x2_pca_axis, x3_pca_axis, x4_pca_axis = 3 * components.T
```

```python
plt.clf()

plt.scatter(x1_pca_axis, x2_pca_axis)
plt.scatter(x1_pca_axis, x3_pca_axis)
plt.scatter(x1_pca_axis, x4_pca_axis)
plt.scatter(x2_pca_axis, x3_pca_axis)
plt.scatter(x2_pca_axis, x4_pca_axis)
plt.scatter(x3_pca_axis, x4_pca_axis)
```
