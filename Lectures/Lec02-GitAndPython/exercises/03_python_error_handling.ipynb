# Python Error Handling

A short reference guide to how we best handle errors in Python, and recommended patterns. We'll also do our best to document some common errors we're seeing in class, and have an answer section on what they mean and how to solve.

## Class Notes

### What's a Python Error?

In spoken language or learning an instrument, we can generally get away with error (say, poor intonation, incorrect technique, grammar mistakes) and others will be able to (for the most part) understand us, or with the instrument, we can still make music.

Unfortunately, with programming languages, everything ultimately needs to be as perfect as possible, especially when working with our own code.

### Where do errors come from?

The easiest way (I've found) to examine an ErrorClass is to make one! Let's try to "catch" the error:

````python
try:
    print hello
except Exception as e:
    print type(e)
    print e
````

Now we've identified that NameError comes from the exceptions module:

````python
import exceptions
help(exceptions)
````

From here, we can actually look up documentation on an error. Documentation is one thing that makes python _great_.

[Documentation for NameError](https://docs.python.org/2/library/exceptions.html#exceptions.NameError)

[Documentation for IOError](https://docs.python.org/2/library/exceptions.html#exceptions.IOError)

### Steps to resolving errors

#### 1. Triple Check your syntax!
More often than not, we have bugs because of syntax. This would be the equivelent of poor grammer, and Python will usually report back a SyntaxError.

1. Check your syntax pairs: `[]`, `{}`, `()`, `''`, `""`, etc. Are they all lined up correctly?
2. Check indentation and colons (`:`). Are any missing on loops, if/else, etc?
3. Check phrasing: is it actally python code you're running?

#### 2. Is something not defined?
1. This might occur because you deleted code in ipython notebook that already ran. Just because you deleted it does not mean it was removed from memory. You might need to backtrack a bit.
2. Are your libraries and modules correctly imported?

#### 3. All other issues, but your own code:
1. Objectively (especially from a data exploration POV), you should not have ANY errors in your code. Actual errors more often occur when working with _another's_ code.
2. Read your code and the error carefully. What is the error explaining?
3. Find the documentation on the error and read that. Does this help figure out what is occurring?
4. Consider searching for the error on google and stackoverflow.
    * Remove any references to your variables, columns, data. For example, if the error was `NameError: 'hello' is not defined`, then search 'NameError not defined' in Google.
    * Observe if it is a common problem (like it is tagged as a bug in github) or common user error (there are duplicate questions on stack overflow).
    * If you can't find a result, make a dummy username on stackoverflow and post some sample code that generated the error and as about what it means.
5. Do not use try/catch when you are doing all the data manipulation in a controlled environment. This is bad practice: find a smarter way to handle your errors:

**DO NOT DO**
```python
try:
    print hello
except:
    hello = 'defining name hello now'
    print hello
```

**DO**
Just define it first:

```python
hello = 'already defined'
print hello
```

Or use control flow to evaluate information before using it:

```python
if 'hello' in locals():
    print hello
```

#### 4. All other issues, but someone else's code (primarily APIs, not modules)
1. If you know what the data structure is expected to look like (given xml or json), consider testing for truth that variables exist. If they don't, then don't run code that depends on them
2. If you're expecting a response but do not know what the data will look like (for example, if an API sends back JSON with an error code), use try/catch. 
3. If the code you're working with is "extra" and the final result doesn't necessarily depend on data being present, consider using try/catch. 
4. If the API has documentation, read it to learn about what error codes are associated with what meaning.
