## Modeling rent prices based on income

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from scipy.stats import linregress

%matplotlib inline

x = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
y = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])

fig = plt.figure(1, figsize=(20, 10), dpi=5, facecolor='r', edgecolor='g')
ax = fig.add_subplot(1,1,1)
fig.add_subplot

x = np.linspace(0, 16, 800)
y = np.sin(x)

ax.plot(x, y, 'b-', linewidth=2, label=r'$y=\sin(x)$')

plt.title('Housing Prices as a Funciton of Income')
plt.xlabel('Monthly Income (Dollars)')
plt.ylabel('Monthly Rent expenditure (Dollars)')
plt.grid(True)
plt.figtext(0.995, 0.01, 'Footnote', ha='right', va='bottom')
plt.legend(loc='best', framealpha=0.5, prop={'size': 'small'})

fig.savefig('filename.png', dpi=125)
```

```python
# --- get the Figure and Axes all at once 
fig, ax = plt.subplots(figsize=(20,12))

# --- plot some lines
N = 8 # the number of lines we will plot 
styles = ['-', '--', '-.', ':']
markers = list('+ox^psDv')  # Equivalent to ['+', 'o', 'x', '^', 'p', 's', 'D', 'v']
x = np.linspace(0, 100, 20)

for i in range(N): # add line-by-line 
    y = x + x/5*i + i
    s = styles[i % len(styles)]
    m = markers[i % len(markers)] 
    ax.plot(x, y, label='Line '+str(i+1)+' '+s+m, marker=m, linewidth=2, linestyle=s)
    
# --- add grid, legend, title and save 
ax.grid(True)
ax.set_xlabel('Income')
ax.set_ylabel('Income')
ax.legend(loc='best', prop={'size':'large'}) 
fig.suptitle('A Simple Line Plot') 
fig.savefig('filename.png', dpi=125)
```
