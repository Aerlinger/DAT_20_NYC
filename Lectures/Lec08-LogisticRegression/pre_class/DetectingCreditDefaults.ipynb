```python
## This function is used to plot the decision boundary in two dimensions. We haven't covered some of this stuff yet, 
# and while useful, focus on the rest of this code in this notebook before reviewing this function.
def plot_decision_boundary(est, x_1, x_2, ax=None, threshold=0.0, contourf=False):
    """Plots the decision surface of ``est`` on features ``x1`` and ``x2``. """
    xx1, xx2 = np.meshgrid(np.linspace(x_1.min(), x_1.max(), 100), 
                           np.linspace(x_2.min(), x_2.max(), 100))
    # plot the hyperplane by evaluating the parameters on the grid
    X_pred = np.c_[xx1.ravel(), xx2.ravel()]  # convert 2d grid into seq of points
    if hasattr(est, 'predict_proba'):  # check if ``est`` supports probabilities
        # take probability of positive class
        pred = est.predict_proba(X_pred)[:, 1]
    else:
        pred = est.predict(X_pred)
    Z = pred.reshape((100, 100))  # reshape seq to grid
    if ax is None:
        ax = plt.gca()

    # plot line via contour plot

    if contourf:
        ax.contourf(xx1, xx2, Z, levels=np.linspace(0, 1.0, 10), cmap=plt.cm.RdBu, alpha=0.6)
    ax.contour(xx1, xx2, Z, levels=[threshold], colors='black')
    
    ax.set_xlim((x_1.min(), x_1.max()))
    ax.set_ylim((x_2.min(), x_2.max()))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.colors import ListedColormap

%matplotlib inline

df = pd.read_csv('../../../Data/Default.csv')

print df.columns

# Let's quickly describe the data
df.describe()
```

```python


# It's sometimes useful to use a cross-tab to calculate relationships between columns.
pd.crosstab(df['student'], df['default']).rename(columns={0: 'No', 1: 'Yes'})
```

```python
df.head()
```

*Note that the populations of non-default vs. defaulted credit cards is very asymmetrical. We'll discuss strategies of dealing with these cases in future classes, but for simplicity in this exersise, we'll downsample observations that haven't defaulted.*

```python
# downsample negative cases -- there are many more negatives than positives
indices = np.where(df.default == 0)[0]
rng = np.random.RandomState(13)
rng.shuffle(indices)
n_pos = (df.default == 1).sum()
df = df.drop(df.index[indices[n_pos:]])

df.head()

pd.crosstab(df['student'], df['default']).rename(columns={0: 'No', 1: 'Yes'})
```

```python
# Explore our data by coloring datapoints by whether the account defaulted with balance and income on the axes
plt.figure(figsize=(14, 8))

# scatter plot of balance (x) and income (y)
ax1 = plt.subplot(221, title="Credit defaults by income and balance")
cm_bright = ListedColormap(['#FF0000', '#00FF00'])
ax1.scatter(df.balance, df.income, c=(df.default == 1), cmap=cm_bright)        
ax1.set_xlim((df.balance.min(), df.balance.max()))
ax1.set_ylim((df.income.min(), df.income.max()))
ax1.set_xlabel('balance')
ax1.set_ylabel('income')
ax1.legend(loc='upper right')

# box plots for income
ax2 = plt.subplot(222)
ax2.boxplot([df.income[df.default == 1], df.income[df.default == 0]])
ax2.set_ylim((df.income.min(), df.income.max()))
ax2.set_xticklabels(('Yes', 'No'))
ax2.set_ylabel('income')

# box plots for balance
ax3 = plt.subplot(223)
ax3.boxplot([df.balance[df.default == 0], df.balance[df.default == 1]], vert=0)
ax3.set_xlim((df.balance.min(), df.balance.max()))
ax3.set_yticklabels(('Yes', 'No'))
ax3.set_xlabel('balance')
plt.tight_layout()
```

```python
from sklearn.linear_model import LinearRegression

fig = plt.figure(figsize=(15, 10))

axes = plt.gca()

# get feature/predictor matrix as numpy array
X = df[['balance', 'income']].values

# encode class labels
classes, y = np.unique(df.default.values, return_inverse=True)
y = (y * 2) - 1

# fit OLS regression 
est = LinearRegression(fit_intercept=True, normalize=True)
est.fit(X, y)

axes.scatter(df['balance'], y, alpha=0.05, marker='o', s=80)
axes.set_title("Credit Defaults by Account Balance", fontsize=24)
axes.set_xlabel('Account Balance (USD)', fontsize=15)
axes.set_ylabel('Defaulted', fontsize=15)

xlim = axes.get_xlim()

y_hat = np.multiply(xlim, est.coef_[0]) + est.intercept_

axes.plot(xlim, y_hat, c='r', linestyle=':')
axes.set_xlim(xlim)
```

### How might I use regression to predict whether an account will default or not?

```python
# plot data and decision surface

plt.figure(figsize=(15, 10))
ax = plt.gca()

ax.set_title("Credit Defaults with decision doundary", fontsize=24)
ax.set_xlabel("Account Balance", fontsize=18)
ax.set_ylabel("Income", fontsize=18)

ax.scatter(df.balance, df.income, c=(df.default == 1), cmap=cm_bright, s=150)

plot_decision_boundary(est, X[:, 0], X[:, 1], ax=ax)

p1 = plt.Rectangle((0, 0), 1, 1, fc='r')
p3 = plt.Rectangle((0, 0), 1, 1, fc='#00FF00')
plt.legend((p1, p3), ('Credit Defaulted','No Default'))

plt.show();
```
