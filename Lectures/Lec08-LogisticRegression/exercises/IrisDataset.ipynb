```python
# From Sklearn examples

import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model, datasets

%matplotlib inline

# Step 1: Load the data from sklearn datasets, and define the domain of X and Y
iris = datasets.load_iris()
  
X = iris.data[:, :2]
Y = iris.target

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
```

```python
# Step 2: Build the model using: linear_model.LogisticRegression
inverse_regularization = 1e10

logreg = linear_model.LogisticRegression(C=1e5)
logreg.fit(X, Y);

print "Score: ", logreg.score(X, Y);

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])

# # Reshape Z from a 1D shape to a 2d plot to plot the decision boundaries
Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(10, 10))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.show()
```

```python
# Applying stronger regularization:
inverse_regularization = 2

logreg = linear_model.LogisticRegression(C=inverse_regularization)
logreg.fit(X, Y);

print "Score: ", logreg.score(X, Y)

Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])

# Reshape Z from a 1D shape to a 2d plot to plot the decision boundaries
Z = Z.reshape(xx.shape)

plt.figure(1, figsize=(10, 10))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.show()
```
