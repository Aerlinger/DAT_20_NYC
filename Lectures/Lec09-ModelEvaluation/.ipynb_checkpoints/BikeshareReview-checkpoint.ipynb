```python
import pandas as pd
import numpy as np
import csv as csv
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import sklearn.linear_model
import sklearn.svm
from datetime import datetime, date, time
from IPython.display import Image
sns.set()

import sklearn.cross_validation

from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression

%matplotlib inline

pd.options.mode.chained_assignment = None

# - instant: record index
# - dteday : date
# - season : season (1:springer, 2:summer, 3:fall, 4:winter)
# - yr : year (0: 2011, 1:2012)
# - mnth : month ( 1 to 12)
# - hr : hour (0 to 23)
# - holiday : weather day is holiday or not (extracted from [Web Link])
# - weekday : day of the week
# - workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
# + weathersit : 
# - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
# - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
# - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
# - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
# - temp : Normalized temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-8, t_max=+39 (only in hourly scale)
# - atemp: Normalized feeling temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-16, t_max=+50 (only in hourly scale)
# - hum: Normalized humidity. The values are divided to 100 (max)
# - windspeed: Normalized wind speed. The values are divided to 67 (max)
# - casual: count of casual users
# - registered: count of registered users
# - cnt: count of total rental bikes including both casual and registered


bikeshare_csv = pd.read_csv('../Data/bikeshare.csv').drop(['instant'], axis=1)

bikeshare = bikeshare_csv.rename(
    columns= {
        'dteday': 'datetime', 
        'weathersit': 'weather', 
        'yr': 'year', 
        'mnth': 'month', 
        'hum': 'humidity', 
        'weekday': 'dayofweek', 
        'hr': 'hour', 
        'cnt': 'count' 
    }
)

bikeshare.describe()
```

```python
bikeshare['hour'].value_counts()
```

```python
# Calculating the rush-hour column

bikeshare['rushhour'] = ((bikeshare['hour'] >= 6) & (bikeshare['hour'] <= 9) | (bikeshare['hour'] >= 16) & (bikeshare['hour'] <= 18)).astype(np.int)
```

```python
#Creating the variable of the medians based on day and hour
#I choose median because it tends to reflect better the typical situation, and minimizes the influence of outliers.
bikeshare['MedianCount-dt']= np.zeros(len(bikeshare)) #dt refers to by day and time of the day
for x in xrange(7):
    for y in xrange(24):
        bikeshare['MedianCount-dt'][(bikeshare.dayofweek==x) &(bikeshare.hour==y)] = np.median(bikeshare['count'][(bikeshare.dayofweek==x) & (bikeshare.hour==y)])
```

```python
bikeshare['MedianCount-dt'].value_counts()
```

```python
#Create new columns for day, month, year, hour
bikeshare['year'] = pd.DatetimeIndex(bikeshare['datetime']).year
bikeshare['month'] = pd.DatetimeIndex(bikeshare['datetime']).month
bikeshare['day'] = pd.DatetimeIndex(bikeshare['datetime']).day
bikeshare['dayofweek'] = pd.DatetimeIndex(bikeshare['datetime']).dayofweek # 0 = Monday to 6 = Sunday
```

```python
ohe = OneHotEncoder(sparse=False)
ohe_encoded = ohe.fit_transform(bikeshare[['weather']])

bikeshare_encoded = pd.concat([bikeshare, pd.DataFrame(ohe_encoded)], axis=1)

bikeshare_encoded.rename(columns={0: 'Sunny', 1: 'Cloudy', 2: 'Snowy', 3: 'Rainy'}, inplace=True)
```

```python
# Train/test split

N = 10885

train_df = bikeshare_encoded.loc[:N, :]
test_df = bikeshare_encoded.loc[(N+1):, :]

# features = ['rushhour', 'season', 'temp', 'humidity', 'workingday', 'hour', 'Sunny', 'Cloudy', 'Snowy', 'Rainy']
features = train_df.columns - ['datetime', 'count', 'registered', 'casual', 'year', 'MedianCount-dt', 'atemp']

X_train = train_df[features]
y_train = train_df[['MedianCount-dt']]

X_test = test_df[features]
y_test = test_df[['MedianCount-dt']]
```

```python
#Rescaling MedianCount-dt
train_df['MedianCount-dt']= train_df['MedianCount-dt']/1000 #So that everytime I predict, I know I should multiply by 1000.

#Definitely keep the stopwatchtime variable (equivalent of a lag variable)
train_df['stopwatchTime'] = range(1,len(train_df)+1) #Scale it
max_stopwatch = max(train_df['stopwatchTime'])
train_df.stopwatchTime = train_df.stopwatchTime/max_stopwatch

#Transforming count to scale
train_df['count']= train_df['count']/1000 #So that everytime I predict, I know I should multiply by 1000.
```

```python
#In our dataframe, we are transforming year into 0 = 2011, 1 = 2012 for both training and test set
train_df.loc[train_df['year']==2011, 'year'] = 0
train_df.loc[train_df['year']==2012, ('year')] = 1
test_df.loc[test_df['year']==2011, ('year')] = 0
test_df.loc[test_df['year']==2012, ('year')] = 1
```

```python
lm = LinearRegression()

model = lm.fit(X_train, y_train)

y_pred_test = model.predict(X_test)
```
