```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python

# # This is how to see your current working directory (cwd).
# import os
# print os.getcwd()

advertising = pd.read_csv('../../Data/Advertising.csv')

advertising = advertising.drop('Unnamed: 0', axis=1)

N = 200

advertising = advertising.head(N)
```

```python
x = advertising['TV']
y = advertising['Sales']

plt.figure(figsize=(18, 9))
axes = plt.gca()

axes.grid(True)
axes.scatter(x, y, c='r', s=150, alpha=1, linewidth=1)

y_mean = y.mean()

axes.set_title('Sales vs. TV Advertisement Spend', fontsize=18)
axes.set_xlabel('TV Advertisement Spend')
axes.set_ylabel('Sales')

axes.axhline(y_mean, c='g', linewidth=1, linestyle='--')

axes.vlines(x, y_mean, y, linestyle=':', alpha=1, colors=['blue'])
```

```python
# Step 2: Deriving beta_0 and beta_1, from a set of points in X and y

from sklearn import linear_model

regressor = linear_model.LinearRegression()

print x.shape

regressor.fit(x[:, np.newaxis], y)

print regressor.coef_
print regressor.intercept_

```

```python
x_min = x.min()
x_max = x.max()

# plt.figure(figsize=(20, 20))
plt.figure()

axes_regression = plt.gca()

# 1. Find y1 and y2 and plot them using the formula below
# print x

y_hat = np.multiply(regressor.coef_, x) + regressor.intercept_

# A plot of my model with points for each value of x
axes_regression.scatter(x, y_hat, c='magenta', marker='+', linewidth=1, s=150)
axes_regression.plot(x, y_hat, c='r')
axes_regression.scatter(x, y, c='r')

axes_regression.grid(True)
# 2. On the same axes overlay the 20 points from the previous cell

axes_regression.vlines(x, y_hat, y, colors=['g'])
```

```python
regressor.score(x[:, np.newaxis], y)
```

```python
residuals = y - y_hat

residuals_of_mean = y - y_mean

# SSR = sum(np.multiply(residuals, residuals))
# SST = sum(np.multiply(residuals_of_mean, residuals_of_mean))

SSR = np.dot(residuals, residuals)
SST = np.dot(residuals_of_mean, residuals_of_mean)

r_squared = 1 - float(SSR)/float(SST)

print r_squared
```
