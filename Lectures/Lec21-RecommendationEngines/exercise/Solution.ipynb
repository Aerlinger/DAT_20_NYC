```python
def jaccard(first, second):
  first = set(first)
  second = set(second)
  
  # the line below should be changed
  union = first.union(second)
  intersection = first.intersection(second)

  return len(intersection) / float(len(union))
```

```python
jaccard(brandsfor['83065'], brandsfor['82983'])
```
